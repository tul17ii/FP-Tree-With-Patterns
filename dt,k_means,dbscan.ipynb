```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Step 1: Create the DataFrame
data = {
    'age': ['youth', 'youth', 'middle_aged', 'senior', 'senior', 'middle_aged', 'youth', 'youth', 'senior', 'senior', 'youth', 'middle_aged', 'middle_aged', 'senior'],
    'income': ['high', 'high', 'high', 'low', 'low', 'low', 'medium', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent', 'fair', 'excellent', 'fair', 'excellent', 'excellent', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'no']
}

df = pd.DataFrame(data)

# Step 2: Convert categorical columns into numeric using LabelEncoder
le = LabelEncoder()

df['age'] = le.fit_transform(df['age'])
df['income'] = le.fit_transform(df['income'])
df['student'] = le.fit_transform(df['student'])
df['credit_rating'] = le.fit_transform(df['credit_rating'])
df['buys_computer'] = le.fit_transform(df['buys_computer'])

# Step 3: Define features (X) and target (y)
X = df.drop(columns=['buys_computer'])  # Features
y = df['buys_computer']  # Target

# Step 4: Build a Decision Tree Classifier
dt = DecisionTreeClassifier()
dt.fit(X, y)

# Step 5: Predict for the given samples
# Sample 1: (age = middle_aged, income = low, student = yes, credit_rating = excellent)
sample1 = pd.DataFrame([[2, 0, 1, 1]], columns=X.columns)  # Encoded values for the sample
prediction1 = dt.predict(sample1)

# Sample 2: (age = senior, income = high, student = no, credit_rating = excellent)
sample2 = pd.DataFrame([[3, 2, 0, 1]], columns=X.columns)  # Encoded values for the sample
prediction2 = dt.predict(sample2)

# Step 6: Print predictions
print(f"Prediction for Sample 1 (middle_aged, low, yes, excellent): {le.inverse_transform(prediction1)[0]}")
print(f"Prediction for Sample 2 (senior, high, no, excellent): {le.inverse_transform(prediction2)[0]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Given Data Points
data_points = np.array([[3, 11], [8, 4], [5, 8], [5, 11], [10, 7], [14, 9], [20, 6], [22, 8]])

# Step 2: Initial Cluster Centers (Centroids)
initial_centroids = np.array([[3, 11], [5, 11], [20, 6]])

# Step 3: Apply K-Means Algorithm
kmeans = KMeans(n_clusters=3, init=initial_centroids, n_init=1, max_iter=100)
kmeans.fit(data_points)

# Step 4: Get the Cluster Labels and Centroids
labels = kmeans.labels_  # Data points cluster assignments
centroids = kmeans.cluster_centers_  # Final centroids

# Output the labels (cluster assignments) and final centroids
print("Cluster Assignments:", labels)
print("Final Centroids:", centroids)

# Plot the data points and the centroids
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Given data points
data_points = np.array([[3, 11], [8, 4], [5, 8], [5, 11], [10, 7], [14, 9], [20, 6], [22, 8]])

# Initial Cluster Centers
initial_centroids = np.array([[3, 11], [5, 11], [20, 6]])

# Apply K-Means Algorithm
kmeans = KMeans(n_clusters=3, init=initial_centroids, n_init=1, max_iter=100)
kmeans.fit(data_points)

# Cluster labels (which cluster each point belongs to)
labels = kmeans.labels_

# Final centroids
centroids = kmeans.cluster_centers_

# Output the cluster labels and centroids
print("Cluster Labels:", labels)
print("Final Centroids:", centroids)

# Plot the data points and the centroids
plt.figure(figsize=(8, 6))

# Plot the data points with different colors for each cluster
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis', marker='o')

# Plot the initial centroids (in blue)
plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1], color='blue', s=200, marker='x', label='Initial Centroids')

# Plot the final centroids (in red)
plt.scatter(centroids[:, 0], centroids[:, 1], color='red', s=200, marker='X', label='Final Centroids')

plt.title('K-Means Clustering (3 Clusters)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import pairwise_distances
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Given data points
data_points = np.array([[3, 11], [8, 4], [5, 8], [5, 11], [10, 7], [14, 9], [20, 6], [22, 8]])

# Step 2: Apply DBSCAN Clustering
# Define DBSCAN with epsilon=3 and min_samples=2 (based on the data)
dbscan = DBSCAN(eps=3, min_samples=2)
dbscan_labels = dbscan.fit_predict(data_points)

# Step 3: Visualizing DBSCAN clustering result
plt.figure(figsize=(8, 6))
plt.scatter(data_points[:, 0], data_points[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()
plt.show()

# Step 4: K-Medoids Algorithm - We need to find 3 clusters as per the problem statement
# First, we initialize random cluster centers based on the problem description
initial_centroids = np.array([[3, 11], [5, 11], [20, 6]])

# Step 5: K-Medoids using KMeans (For simplicity)
kmeans = KMeans(n_clusters=3, init=initial_centroids, n_init=1)
kmeans.fit(data_points)

# Step 6: Visualizing the K-Medoids result
plt.figure(figsize=(8, 6))
plt.scatter(data_points[:, 0], data_points[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')
plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1], color='red', marker='x', s=200, label='Initial Centroids')
plt.title("K-Medoids Clustering (KMeans for initialization)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

# Step 7: Output the results
print("DBSCAN Labels:", dbscan_labels)
print("K-Medoids Cluster Labels:", kmeans.labels_)
print("K-Medoids Final Centroids:", kmeans.cluster_centers_)
```
